# 24-Hour Input Events Visualization

This notebook creates timeline visualizations for ICU patients showing medications and interventions administered over 24 hours based on MIMIC-III dataset.

## Import required Python libs

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import numpy as np
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
```

## Data Loading with Spark (Dataframes)

```python
def load_patient_data_by_id(icustay_id, data_path):

    spark = SparkSession.builder.appName("PatientFilter").getOrCreate()
    
    input_events = spark.read.csv(f"{data_path}/INPUTEVENTS_MV.csv", header=True, inferSchema=True)
    patient_data = input_events.filter(col("ICUSTAY_ID") == icustay_id)
    
    # Convert to Pandas for plotting
    patient_df = patient_data.toPandas()
    
    spark.stop()
    return patient_df
```

## Creating patient treatment timeline from INPUTEVENTS_MV dataset.

What this section does:
- Data Loading: Uses our efficient function to load only relevant data
- Item Descriptions: Merges with D_ITEMS.csv to get medication names
- Error Handling: Checks if patient exists in the dataset
- Data Integration: Combines input events with descriptive labels
- Visualization: Plots the respective chart for a given patient intensive care unit stay.


```python
def create_24h_patient_timeline(icustay_id, data_path):
    
    print(f"Loading data for ICUSTAY_ID: {icustay_id}")
    patient_data = load_patient_data_by_id(icustay_id, data_path)
    
    # Load item descriptions
    d_items = pd.read_csv(f"{data_path}/D_ITEMS.csv")
    
    if patient_data.empty:
        print(f"No data found for ICUSTAY_ID {icustay_id}")
        return
    
    # Merge with item descriptions
    patient_data = patient_data.merge(d_items[['ITEMID', 'LABEL']], on='ITEMID', how='left')
    
    patient_data['STARTTIME'] = pd.to_datetime(patient_data['STARTTIME'])
    start_time = patient_data['STARTTIME'].min()
    patient_data['hours_from_start'] = (patient_data['STARTTIME'] - start_time).dt.total_seconds() / 3600
    patient_data['day_fraction'] = patient_data['hours_from_start'] / 24
    
    # Filter for first 24 hours only (day_fraction between 0 and 1)
    patient_data = patient_data[patient_data['day_fraction'] <= 1.0]
    
    numeric_data = patient_data[patient_data['AMOUNT'].notna() & 
                               (patient_data['AMOUNT'] > 0)].copy()
    
    if numeric_data.empty:
        print("No numeric values found for this patient in the first 24 hours")
        return
    

    # Create the plot
    plt.figure(figsize=(14, 8))
    unique_items = numeric_data['ITEMID'].unique()
    colors = plt.cm.Set3(np.linspace(0, 1, len(unique_items)))
    
    # Plot each item type with different color
    for i, item_id in enumerate(unique_items):
        item_data = numeric_data[numeric_data['ITEMID'] == item_id]
        label_name = item_data['LABEL'].iloc[0] if not pd.isna(item_data['LABEL'].iloc[0]) else f"Item {item_id}"
        
        plt.scatter(item_data['day_fraction'], 
                   item_data['AMOUNT'],  # Use AMOUNT instead of VALUENUM
                   c=[colors[i]], 
                   label=f"{item_id}: {label_name}"[:40],  # Truncate long labels
                   alpha=0.7,
                   s=50)
    
    # Customize the plot to match assignment example
    plt.xlabel('Fraction of Day (0.25 = 6 hours, 0.5 = 12 hours, etc.)')
    plt.ylabel('Amount/Quantity Given')
    plt.title(f'24-Hour Input Events Timeline - ICUSTAY_ID: {icustay_id}')
    
    # Add grid for better readability
    plt.grid(True, alpha=0.3)
    
    # Set x-axis limits and ticks
    plt.xlim(0, 1)
    plt.xticks([0, 0.25, 0.5, 0.75, 1.0], 
               ['0h\n(Start)', '6h\n(0.25)', '12h\n(0.5)', '18h\n(0.75)', '24h\n(1.0)'])
    
    # Place legend on the right side (like assignment example)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
    
    plt.tight_layout()
    
    # Save image
    plt.savefig(f'input_events_timeline_{icustay_id}.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print(f"\nðŸ“Š Summary for ICUSTAY_ID {icustay_id}:")
    print(f"   â€¢ Time period: {start_time.strftime('%Y-%m-%d %H:%M')} to {(start_time + timedelta(hours=24)).strftime('%Y-%m-%d %H:%M')}")
    print(f"   â€¢ Total events in 24h: {len(numeric_data)}")
    print(f"   â€¢ Unique items: {len(unique_items)}")
    print(f"   â€¢ Most frequent item: {numeric_data['ITEMID'].mode().iloc[0]} ({numeric_data[numeric_data['ITEMID'] == numeric_data['ITEMID'].mode().iloc[0]]['LABEL'].iloc[0]})")
    
    return numeric_data
```

# 3. Execute the Function and Visualize the Data

```python
data = create_24h_patient_timeline(200072, "./mimic-db/")
```
